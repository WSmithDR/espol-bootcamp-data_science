```python
from transformers import pipeline

def analizar_sentimientos(textos):
    """
    Realiza análisis de sentimientos en español.

    Args:
        textos (list of str): Lista de textos para analizar.

    Returns:
        list of dict: Lista de resultados con las etiquetas y puntuaciones.
    """
    # Cargar el pipeline de análisis de sentimientos con un modelo multilingüe
    analizador_sentimientos = pipeline(
        "sentiment-analysis",
        model="nlptown/bert-base-multilingual-uncased-sentiment"
    )

    # Analizar cada texto
    resultados = analizador_sentimientos(textos)

    return resultados
```

```python
textos = [
    "la atencion en este sitio es mala ya quiero volver a casa",
    "eficiente y comodo",
    "Cálida bienvenida, personal eficiente",
    "este sitio es muy bueno y la atencion es de lo mejor",
    "la atencion en este sitio es mala ya quiero volver a casa",
    "espero nunca jamas volver a un lugar tan despreciable y triste como este",
    "Me encanta este producto, es realmente fantástico.",
    "Estoy muy decepcionado con el servicio.",
    "La película fue bastante aburrida.",
    "¡Qué experiencia tan increible!",
    "No volveré a comprar aqui, pésima atención al cliente."
]

# Realizar análisis de sentimientos
resultados = analizar_sentimientos(textos)
resultados
```

```python
lista_tweets_filtrados
```

```python
lista_tweets_filtrados = [str(tweet) for tweet in lista_tweets_filtrados]
```

```python
analisis_sentimientos_tweets = analizar_sentimientos(lista_tweets_filtrados)
```

```python
analisis_sentimientos_tweets
```

```python
import matplotlib.pyplot as plt
```

```python
df_sentimientos = pd.DataFrame(analisis_sentimientos_tweets)
```

```python
df_sentimientos
```

```python
df_sentimientos.shape
```

```python
conteo_sentimientos = df_sentimientos['label'].value_counts()
```

```python
conteo_sentimientos
```

```python
plt.figure(figsize=(10, 6))
colors = {
    '1 star': 'red',
    '2 stars': 'orange',
    '3 stars': 'yellow',
    '4 stars': 'lightgreen',
    '5 stars': 'green'
}


conteo_sentimientos.plot(kind='bar', color=[colors.get(x, 'blue') for x in conteo_sentimientos.index])


plt.title('Distribución de Sentimientos en los Tweets')
plt.xlabel('Calificación')
plt.ylabel('Cantidad de Tweets')
plt.xticks(rotation=45)


for i, v in enumerate(conteo_sentimientos):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
